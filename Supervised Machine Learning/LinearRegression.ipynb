```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

x --> area_sample = pd.Series([1, 2, 3, 4 , 5])  
y --> price_sample = pd.Series([3, 4, 2, 4 , 5])  

x̄ = mean --> shows the density of the data and the pattern
1+2+3+4+5/5 --> 3

ȳ = mean
3+4+2+4+5/5 --> 3.6

∑(x-x̄) * ∑(y-ȳ) -->  4

∑(x-x̄)^2 --> 10


slope m = ∑(x-x̄)(y-ȳ)/∑(x-x̄)^2 

slope m = 4/10 --> 0.4  --> coef_

ȳ = mx̄ + c
3.6 = 0.4*3 + c

c = 2.4 --> intercept_


```python
#adding x and  y axis
plt.xlabel('area')
plt.ylabel('price')

area_sample = pd.Series([1, 2, 3, 4 , 5])  # Example feature (1D)
# price_sample = pd.Series([10, 25, 30, 48])  # Example target (1D)
price_sample = pd.Series([3, 4, 2, 4 , 5])  # Example target (1D)

#create the graph
plt.scatter(area_sample,price_sample,color='blue')
```

```python
x = [1, 2, 3, 4 , 5]
y = [3, 4, 2, 4 , 5]


x̄ = np.mean(x)
ȳ = np.mean(y)


numerator = 0
denominator = 0

n = len(x)


for i in range(len(x)):
    numerator += (x[i] - x̄) * (y[i] - ȳ)  # Added `*` here
    denominator += (x[i] - x̄) ** 2


m = numerator / denominator
c = ȳ - m * x̄

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
```

```python
max_x = np.max(x)
min_x =np.min(x)

X = np.linspace(min_x,max_x,100)

Y = m*X + c

plt.plot(X,Y)
plt.scatter(x,y)
plt.xlabel("independent variable")
plt.ylabel("dependent variable")

```

```python
# creating a regression model

lreg = linear_model.LinearRegression()

# lreg.fit(x,y)

lreg.fit(area_sample,price_sample)
```

```python
lreg.predict([[5]])
```

```python
#y = mx + c --> lets find m (coef_)
lreg.coef_
```

```python
#y = mx + c --> lets find c(intercept_)
lreg.intercept_
```

```python
#checking regression performance 
# R^2 (coefficient of determination)--> statistical measure of how close the data are to the fitted regression line 
#Relationship between actual value and predicted value 
# R^2 = 1-(∑(yp-ȳ)^2)/(∑(y-ȳ)^2)
ss_t=0
ss_r=0
for i in range(n):
    y_pred=m*X[i]+c
    ss_t += (y[i] - ȳ)**2
    ss_r += (y[i] - y_pred)**2
    
r2 = 1-ss_t/ss_r
print(r2)
    
    
    
```
