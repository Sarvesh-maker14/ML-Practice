```python
import pandas as pd
import numpy as np 
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("extracted_data.csv")
df
```

Price = b0*Area + b1*Age + b2

b0,b1 --> coef
b2 --> Intercept


x --> Area,Age
y --> Price

```python
lr = linear_model.LinearRegression()
lr.fit(df.drop('Price',axis='columns'),df.Price) 
```

```python
lr.coef_ # b0 and b1
```

```python
lr.intercept_
```

```python
#find home with 200 sqr ft area , 60 year old 
lr.predict([[200,60]])
```

How to check the relation between independent variables and dependent variable --> data corelation (r)

1.postive corelation --> x increases y increases --> r --> 1
2.negative corelation --> x increases y decreases --> r --> -1

Price = b0*Area + b1*Age + b2

r(coefficent of relation) =  ∑(x-x̄)(y-ȳ)/(√ ((x-x̄))*(√(y-ȳ)))
